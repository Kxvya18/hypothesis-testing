```python
from scipy.stats import binom
#f you have a scenario where you flip a coin 10 times and want to find the probability of getting 7 or fewer heads, you would use the binom.cdf function.
p=1-binom.cdf(56,100,0.5) #5%
#binom.cdf(k, n, p) k-->successes  n--> trials  p-->significance level
print(str( round(p*100,1) ) +'%')            
```

```python
#The PPF is the inverse of the CDF. Given a probability (p),
#the PPF returns the value (x) such that the probability of the random variable being less than or equal to (x) is (p).
print(binom.ppf(0.95,100,0.5)+1)
```
